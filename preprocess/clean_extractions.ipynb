```python
import pandas as pd

extracted_descriptions = pd.read_json("llm_course_descriptions.jsonl", lines=True)
```

```python
threshold = 9
for l in extracted_descriptions['response']:
    if (type(l) != dict): print(type(l))
    else: 
        if len(l['perceived_prerequisites']) > threshold: print(l['perceived_prerequisites'])
        if len(l['contents']) > threshold: print(l['contents'])
        if len(l['outcomes']) > threshold: print(l['outcomes'])
```

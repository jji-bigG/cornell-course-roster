# verify the schema connection

```python
import os

current_dir = os.path.abspath("")

def init_db(
    db_path="roster_reviews.sqlite.db",
    schema_path=os.path.join(current_dir, "schema.sql"),
):
    conn = sqlite3.connect(db_path)
    with open(schema_path, "r") as f:
        conn.executescript(f.read())
    conn.commit()
    return conn

conn = init_db()
cur = conn.cursor()

tables = cur.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
print("Tables in the database:", tables)

# Check if the 'courses' table exists and display its schema
if ("courses",) in tables:
    schema = cur.execute("PRAGMA table_info(courses);").fetchall()
    print("\nSchema of the 'courses' table:", schema)

    # Query to check if there are FA14 courses
    cur.execute('SELECT * FROM courses WHERE semester="FA14"')
    rows = cur.fetchall()
    print("\nFA14 Courses:", rows)
else:
    print("\nThe 'courses' table does not exist in the database.")

conn.close()
```
